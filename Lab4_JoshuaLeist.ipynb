```python
print("Welcome to Pie Hard!")

crusts = ["Hand Tossed", "Pan", "Thin", "Crispy"]
print("Let's build you a pi. First, let's pick a crust. We have the following available: ")
for i, crust in enumerate(crusts, 1):
    print(f"{i}. {crust}")
crust_selection = int(input("What type of crust would you like? Enter 1, 2, 3, or 4: "))
selected_crust = crusts[crust_selection - 1]
print(f"You selected {selected_crust} crust good choice!\n")

sizes = ["8 inch", "12 inch", "20 inch"]
print("Next let's pick the size of your pi. We have the following available: ")
for i, size in enumerate(sizes, 1):
    print(f"{i}. {size}")
size_selection = int(input("What size pizza would you like? Enter 1, 2, or 3: "))
selected_size = sizes[size_selection -1]
print(f"You selected a {selected_size} pizza.\n")

available_meat_toppings = ["sausage", "pepperoni", "ham", "bacon"]
available_veg_toppings = ["onions", "jalapenos", "black olives", "green peppers", "green chile", "mushrooms", "green olives"]
req_toppings = []
num_toppings = int(input("How many toppings do you want to add?"))
for in range(num_toppings):
    toppings = input("Please enter a topping: ").lower()
    if toppings in available_meat_toppings or toppings in available_veg_toppings:
        print(f"We have added {toppings} to your pizza.\n")
        req_toppings.append(toppings)
    else:
        print(f"I'm sorry, we are out of {toppings} at the moment.\n")

subtotal = 0
total = 0
taxes = 0
if selected_size == "8 inch":
    subtotal = 10
elif selected_size == "12 inch":
    subtotal = 12
elif selected_size == "20 inch":
    subtotal = 24

if selected_crust == "Hand Tossed":
    subtotal += 2.12

for topping in req_toppings:
    if topping in available_meat_toppings:
        subtotal += 1.23
    elif topping in available_veg_toppings:
        subtotal += 0.75

taxes = subtotal * 0.0815
total = subtotal + taxes

print("You Ordered: ")
print(f"{selected_size} {selected_crust} pizza")
print("With the selected toppings: ")
for topping in req_toppings:
    print(topping)
    
print(f"\nSubtotal: ${subtotal:.2f}")
print(f"Taxes: ${taxes:.2f}")
print(f"Total: --- ${total:.2f}")
```
